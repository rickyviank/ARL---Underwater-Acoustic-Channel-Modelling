```python
import arlpy.uwapm as pm
import arlpy.plot as plt
import numpy as np
```

```python
pm.models()
print(pm.models())
```

```python
surface = np.array([[r, 0.5+0.5*np.sin(2*np.pi*0.005*r)] for r in np.linspace(0,1000,1001)])

env = pm.create_env2d(rx_range = 100, depth = 100, rx_depth = 9, tx_depth = 7, frequency = 5000, surface = surface)
```

```python
pm.print_env(env)
```

```python
tloss = pm.compute_transmission_loss(env)
pm.plot_transmission_loss(tloss, env = env, clim = [-60,-30], width = 900)
```

```python
arrivals = pm.compute_arrivals(env)
pm.plot_arrivals(arrivals)
ir = pm.arrivals_to_impulse_response(arrivals[arrivals.surface_bounces<2], fs = 192000)
```

```python
plt.plot(np.abs(ir))
```

```python
pm.compute_arrivals(env)
```

```python
rays = pm.compute_eigenrays(env)
pm.plot_rays(rays, env=env)
```
